# <u>DATA RETRIEVAL</u>

### Setting file name

```python
file_name = 'U1S1.TXT'
```

### Read file contents

```python
f = open(file_name, 'r')
contents = f.read()
print(contents)
```

### Setting number of items and modifying contents

```python
number_of_tuples = int(contents.split('\n')[0])
number_of_tuples
```

```python
contents = contents.split('\n')[1:]
contents
```

```python
type(contents)
```

```python
for c in contents:
    c = c.split(' ')
    for x in c:
        x = int(x)
    print(c)
# print(contents)
```
