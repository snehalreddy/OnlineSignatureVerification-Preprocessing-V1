# <u>DATA RETRIEVAL</u>

### Setting file name

```python
file_name = 'U1S1.TXT'
```

### Read file contents and setting up the pandas dataframe

```python
f = open(file_name, 'r')
contents = f.read()
print(contents)
```

### Setting number of items and modifying contents

```python
number_of_tuples = int(contents.split('\n')[0])
number_of_tuples
```

```python
contents = contents.split('\n')[1:-1]
len(contents)
```

```python
type(contents)
```

```python
tempList = []
for c in contents:
    tempList.append(list(map(int, c.split())))
contents = tempList
```

```python
contents
```

### SETTING X, Y, TS, T FOR X-COORDINATE, Y-COORDINATE, TIMESTAMP, TOUCH

```python
X = []
Y = []
TS = []
T = []
AZ = []
AL = []
P = []
```

```python
for c in contents:
    X.append(c[0])
    Y.append(c[1])
    TS.append(c[2])
    T.append(c[3])
    AZ.append(c[4])
    AL.append(c[5])
    P.append(c[6])
```

```python
print('X:\n', X, '\n')
print('Y:\n', Y, '\n')
print('TS:\n', TS, '\n')
print('T:\n', T, '\n')
print('AZ:\n', AZ, '\n')
print('AL:\n', AL, '\n')
print('P:\n', P, '\n')
```

## ATTRIBUTES CALCULATION
### VELOCITY

```python
distance = 0
time = (TS[-1] - TS[0])/1000
```

##### DISTANCE CALCULATION

```python
for i in range(len(X)-1):
        distance += ((X[i+1] - X[i])**2 + (Y[i+1] - Y[i])**2)**0.5
```

```python
print(distance)
print(time)
```

```python
velocity = distance/time
```
